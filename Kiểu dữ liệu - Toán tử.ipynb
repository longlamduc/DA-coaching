## Toán tử quan hệ

```python
# variable
x = 5
y = 4
z = 9
```

```python
x > y
```

```python
x + y >= z
```

```python
x ** 2 <= z
```

```python
(x ** 2 - z) ** 0.5 == y
```

```python
x + y + (x < y) -1 <= z
```

```python
x * y - z <= (z ** 2 + 40) ** 0.5
```

## Toán tử logic

```python
x = 5
y = 4
z = 9
```

```python
x > y and z > x
```

```python
x + y >= z and z ** 0.5 <= y
```

```python
z - y > x or z - x > y
```

```python
not(z - y > x or z - x > y)
```

```python
not(z - y > x or z - x + 1 >= x)
```
